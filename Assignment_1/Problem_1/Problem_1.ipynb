```
from keras.datasets import fashion_mnist
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import matplotlib.image as mpimg
import numpy as np
```

```
(train_data, train_labels), (test_data, test_labels) = fashion_mnist.load_data()
```

```
unique_list_of_labels = []

#name of products at each label
Names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle Boot"]

#find unique labels of given dataset
for i in train_labels:
  if i not in unique_list_of_labels:
    unique_list_of_labels.append(i)

#sort the labels
unique_list_of_labels.sort()

#plot the images corresponding to given label
fig, axarr = plt.subplots(1,len(unique_list_of_labels),figsize=(20,10))
for i in range (0,len(unique_list_of_labels)):
  indx=train_labels.tolist().index(unique_list_of_labels[i])
  axarr[i].imshow(train_data[indx]) 
  axarr[i].set_title(Names[unique_list_of_labels[i]])
```
